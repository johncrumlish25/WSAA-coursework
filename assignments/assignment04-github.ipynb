### Assignment 4 - Github
### Program will replace all the instances of "Andrew" with "John" in text.txt
### Author: John Crumlish

```python
import os

file_path = "text.txt"   
my_name = "John"       


# Read file
with open(file_path, "r") as file:
    content = file.read()

# Replace "Andrew" with "John"
content = content.replace("Andrew", my_name)

# Write updated content back
with open(file_path, "w") as file:
    file.write(content)

# Commit and push to Github
os.system(f"git add {file_path}")
os.system('git commit -m "Replaced Andrew with my name"')
os.system("git push")

print("Done: text.txt updated and pushed.")
```
