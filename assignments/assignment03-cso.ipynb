### This program retrieves the dataset "exchequer account (historical series)" from the CSO, and stores it into a file called "cso.json"
### Author: John Crumlish

```python
import requests
import json

# CSO dataset URL 
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en"

# Send a GET request to retrieve the dataset
response = requests.get(url)

# Convert response to JSON and save to file
with open("cso.json", "w") as file:
    json.dump(response.json(), file, indent=4)

print("Data saved to cso.json")
```
